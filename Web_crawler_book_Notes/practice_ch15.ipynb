```python
import bs4,requests,os

url='http://www.xzw.com/'
url_1=url+'fortune/'
headers={'user-agent':'Mozilla/5.0 (Windows NT 6.2; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.60 Safari/537.36'}
html=requests.get(url_1,headers=headers)
obj_soup=bs4.BeautifulSoup(html.text,'lxml')

img_list=[]

obj_today=obj_soup.find('div',id='list')
obj_id=obj_soup.find('div','alb').find_all('div','al')

print(obj_today.h1.text.strip())
for id_ in obj_id:
    obj_title=id_.dd.strong.text.strip()   # 星座
    obj_birthday=id_.dd.small.text.strip()   # 日期
    obj_txt=id_.dd.p.text.strip()  # 內文
    obj_img=url+id_.dt.a.img['src']
    print('星座名稱:',obj_title)
    print('出生日期:',obj_birthday)
    print('整體運勢:',obj_txt)
    print('照片網址:',obj_img)
    img_list.append(obj_img)

img_dir='img_dir'
if os.path.exists(img_dir)==False:
    os.mkdir(img_dir)

for img_l in img_list:
    try:
        img_get=requests.get(img_l)
        img_file=open(os.path.join(img_dir,os.path.basename(img_l)),'wb')
        # 把img_get分成img_f小份 並用img_file慢慢寫入
        for img_f in img_get.iter_content(5000):
            img_file.write(img_f)
        img_file.close()
        print(img_l,'下載成功')
    except Exception as err:
        print('下載失敗')
print('全部下載完成')
```
