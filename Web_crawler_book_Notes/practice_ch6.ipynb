```python
# 使用md5()與sha1()計算中英文哈希值
# https://sectools.tw/hashlib/ 可參考

import hashlib

data_E=hashlib.md5()
data_C=hashlib.md5()
txts_E=b'Shan Pei Wang'
txts_C='王山珮'
data_E.update(txts_E)
data_C.update(txts_C.encode('utf-8')) # 中文就要給encode('utf-8')

print('md5()將英文數據轉成哈希值:',data_E.digest())
print('md5()將英文數據轉成16進位哈希值:',data_E.hexdigest())
print(type(data_E))
print(type(data_E.digest()))
print(type(data_E.hexdigest()))

print('\nmd5()將中文數據轉成哈希值:',data_C.digest())
print('md5()將中文數據轉成16進位哈希值:',data_C.hexdigest())
print(type(data_C))
print(type(data_C.digest()))
print(type(data_C.hexdigest()))

# sha1()常被用於數位簽章
data_E=hashlib.sha1()
data_C=hashlib.sha1()
data_E.update(txts_E)
data_C.update(txts_C.encode('utf-8')) # 中文就要給encode('utf-8')

print('\nsha1()將英文數據轉成哈希值:',data_E.digest())
print('sha1()將英文數據轉成16進位哈希值:',data_E.hexdigest())
print(type(data_E))
print(type(data_E.hexdigest()))

print('\nsha1()將中文數據轉成哈希值:',data_C.digest())
print('sha1()將中文數據轉成16進位哈希值:',data_C.hexdigest())
print(type(data_C))
print(type(data_C.hexdigest()))

print('\n列出你所使用作業系統平台可使用的hash:',hashlib.algorithms_available)
print('\n列出跨作業系統平台可使用的hash:',hashlib.algorithms_guaranteed)
```
