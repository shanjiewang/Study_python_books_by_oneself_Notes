```python
# 下載臉書照片 課本方式
import facebook
import os,shutil,requests

token='EAAFtie9qUx4BAFZCcl09mWR5pJdSFb3lvNsyzz3Ckokzgt7ZCno3vuQ35llFZChZCH63RmlmCBc4aYWORKQnZClJPP88nZALalQROfVcuxhRzxw28CeQfs5JVrVAnOxOUdPo9ZBXEAlMYEZCzQnVUKr5edRd8OjxxuOiAAMJ59e6IpZBZAbipDHc97dADIdTenbD74obUmZCbm8QYc7sKZCZCZAfKg'
graph=facebook.GraphAPI(access_token=token,version='3.1')

mypic=graph.get_connections(id='me',connection_name='photos?fields=images')
pics=mypic['data']
#print(pic)

pic_dir='out_19'
if os.path.exists(pic_dir)==False:
    os.mkdir(pic_dir)
n=0
for p in pics:
    pic_list=p['images']
    #print(pic_list)
    n+=1
    if n>10:
        break
    for pict in pic_list:
        pic_url=pict['source']
        pic_name=pic_url.split('/')[-1].split('?')[0]   # split('/')[-1]就是用'/'分開後的倒數第一個
        dst=open(pic_dir+'/'+pic_name,'wb')
        #print(dst)
        try:
            fig=requests.get(pic_url,stream=True)  # stream=True就是強制解壓縮
            # shutil.copyfileobj該函式將檔案複製到目標路徑或檔案的物件上。如果目標物件是檔案物件，則需要直接呼叫 copyfileobj() 後關閉。緩衝區大小用於指定緩衝區的長度。它是指在複製的過程中，記憶體中保留的位元組數。
            shutil.copyfileobj(fig.raw,dst) # 使用Python读取raw格式图像并显示 
            dst.close()
            print('下載成功')
        except:
            print('下載失敗')
print('所有照片下載完成')
```
