```python
# 列出ptt目前頁面文章數量.標題.姓名.超連結等等
# ptt瀏覽量大 所以文章數量會不斷變動

import requests,bs4

url='https://www.ptt.cc/bbs/Gossiping/index.html'
html=requests.get(url,cookies={'over18':'1'})  # cookies={'over18':'1'}設這參數表示滿18歲可以直接進入
obj=bs4.BeautifulSoup(html.text,'lxml')
print(type(obj))

articles_lists=[]  
articles_counts=0  # 文章數量從0計算

ptt_divs=obj.find_all('div','r-ent')
print(len(ptt_divs))

for ptt in ptt_divs:
    # 判斷div class="r-ent"裡要有a標籤才繼續
    if ptt.find('a'):
        # 用find找出需要的物件
        title=ptt.find('a').text    # 超連結標題
        author=ptt.find('div','author').text   # 作者姓名
        href=ptt.find('a')['href']   # 超連結網址
        # 再用append建成字典lists
        articles_lists.append({'title':title,'author':author,'href':href})   
    
for articles in articles_lists:
    articles_counts+=1
    print(articles_counts,':',articles['title'])

print('本頁文章數量:',articles_counts)
```

```python
# 增加推文數量 自己的邏輯寫法

import requests,bs4

url='https://www.ptt.cc/bbs/Gossiping/index.html'
html=requests.get(url,cookies={'over18':'1'})  # cookies={'over18':'1'}設這參數表示滿18歲可以直接進入
obj=bs4.BeautifulSoup(html.text,'lxml')
print(type(obj))

articles_lists=[]  
articles_counts=0  # 文章數量從0計算

ptt_divs=obj.find_all('div','r-ent')
print(len(ptt_divs))

for ptt in ptt_divs:
    # 判斷div class="r-ent"裡要有a標籤才繼續
    if ptt.find('a'):
        # 用find找出需要的物件
        title=ptt.find('a').text    # 超連結標題
        author=ptt.find('div','author').text   # 作者姓名
        href=ptt.find('a')['href']   # 超連結網址
        date=ptt.find('div','date').text   # 發文日期
        push_num=ptt.find('div','nrec').text   # 推文數量
        
        # 先用if一次解決push_num問題
        if push_num=='':
            push_num=0
        # str.startswith(str, beg=0,end=len(string))方法檢查字符串是否以str開始 任選限製匹配與給定索引的開始和結束
        elif push_num.startswith('X'):
            push_num='噓>推'
        elif push_num=='爆':
            push_num='推>100'
        else:
            push_num=int(push_num)

        # 再用append建成字典lists
        articles_lists.append({'title':title,'author':author,'href':href,'date':date,'push_num':push_num})   

#print(articles_lists)
# 最低貼文數
push_min=10
for articles in articles_lists:
    articles_counts+=1
    
    # 假如articles['push_num']是文字型態 並且值大於10 就印出
    if type(articles['push_num'])==int and articles['push_num']>=push_min:  
        print('文章編號:',articles_counts)
        print('文章標題:',articles['title'])
        print('文章作者:',articles['author'])
        print('文章網址:',articles['href'])
        print('文章網址:',articles['date'])
        print('推文數量:',articles['push_num'],'\n')
    else:
        print('文章編號:',articles_counts)
        print('未大於10推文數量:',articles['push_num'],'\n')

print('本頁文章數量:',articles_counts)
```

```python
# 自己做 透過前一頁取得所有頁數 然後透過for迴圈每頁-1取得需要的內容成功

import requests,bs4,re

url='https://www.ptt.cc/bbs/Gossiping/index'
h='.html'
html=requests.get(url+h,cookies={'over18':'1'})
obj=bs4.BeautifulSoup(html.text,'lxml')
# print(obj)

div_page=obj.find('div','btn-group btn-group-paging')
front_page=div_page.find_all('a')[1]['href']  # 找到上頁標籤索引 並取得超連結

print(front_page)
dd=front_page.split('.')
ff=dd[0][20:len(dd[0])]
ff=int(ff)

def all_page(all_obj):

    articles_lists=[]  
    articles_counts=0  # 文章數量從0計算

    ptt_divs=all_obj.find_all('div','r-ent')
    print(len(ptt_divs))

    for ptt in ptt_divs:
        # 判斷div class="r-ent"裡要有a標籤才繼續
        if ptt.find('a'):
            # 用find找出需要的物件
            title=ptt.find('a').text    # 超連結標題
            author=ptt.find('div','author').text   # 作者姓名
            href=ptt.find('a')['href']   # 超連結網址
            date=ptt.find('div','date').text   # 發文日期
            push_num=ptt.find('div','nrec').text   # 推文數量

            # 先用if一次解決push_num問題
            if push_num=='':
                push_num=0
            # str.startswith(str, beg=0,end=len(string))方法檢查字符串是否以str開始 任選限製匹配與給定索引的開始和結束
            elif push_num.startswith('X'):
                push_num='噓>推'
            elif push_num=='爆':
                push_num='推>100'
            else:
                push_num=int(push_num)

            # 再用append建成字典lists
            articles_lists.append({'title':title,'author':author,'href':href,'date':date,'push_num':push_num})   

    #print(articles_lists)
    # 最低貼文數
    push_min=10
    for articles in articles_lists:
        articles_counts+=1

        # 假如articles['push_num']是文字型態 並且值大於10 就印出
        if type(articles['push_num'])==int and articles['push_num']>=push_min:  
            print('文章編號:',articles_counts)
            print('文章標題:',articles['title'])
            print('文章作者:',articles['author'])
            print('文章網址:',articles['href'])
            print('文章日期:',articles['date'])
            print('推文數量:',articles['push_num'],'\n')
        else:
            print('文章編號:',articles_counts)
            print('未大於10推文數量:',articles['push_num'],'\n')

    print('本頁文章數量:',articles_counts)
    
for i in range(ff,1,-1):
    print(url,i,h)
    all_html=requests.get(url+str(i)+h,cookies={'over18':'1'})
    all_obj=bs4.BeautifulSoup(all_html.text,'lxml')
    all_page(all_obj)  # 倒退不停重複執行每頁
```

```python
# 更進一步進入貼文超連結

import bs4,requests,os

url='https://www.ptt.cc/'
beauty_page='bbs/beauty/index.html'
html=requests.get(url+beauty_page,cookies={'over18':'1'})
obj=bs4.BeautifulSoup(html.text,'lxml')

ptt_divs=obj.find_all('div','r-ent')
href=ptt_divs[0].find('a')['href']

print('目前的連線網址:',url+href)
page_html=requests.get(url+href,cookies={'over18':'1'})
page_obj=bs4.BeautifulSoup(page_html.text,'lxml')

page_divs=page_obj.find('div',id='main-content')
pushs=page_divs.find_all('div','push')
#print('\ndiv id="main-content"內的物件:',page_divs)
my_list=list(page_divs)  # 直接list()就可以將指定邊謙內所有物件轉成串列
#print('\n直接list()就可以轉成串列:',my_list)

items=page_divs.find_all('div','article-metaline')
for item in items:
    #print(item)
    tag_1=item.find('span','article-meta-tag').text
    tag_2=item.find('span','article-meta-value').text
    print(tag_1,':',tag_2)
    # 需自行找內文索引
print('文章內文:',my_list[4].strip())  # strip()刪除前導和字尾字元

print('\n--- 轉list是因為知道網頁結構 方便直接取索引 以下留言內文區---')
for push in pushs:
    #print(item)
    # 轉list是因為知道網頁結構 方便直接取索引
    push_list=list(push)
    is_push=push_list[0].text
    is_arthor=push_list[1].text
    is_content=push_list[2].text
    is_date=push_list[3].text.strip()
    print('推:',is_push)
    print('作者:',is_arthor)
    print('內文:',is_content)
    print('時間:',is_date)
    
# print('\n---用find找標籤寫法 以下留言內文區---')
# for push in pushs:
#     #print(item)
#     # 用find找標籤寫法
#     is_push=push.find('span','push-tag').text
#     is_arthor=push.find('span','push-userid').text
#     is_content=push.find('span','push-content').text
#     is_date=push.find('span','push-ipdatetime').text
#     print('推:',is_push)
#     print('作者:',is_arthor)
#     print('內文:',is_content)
#     print('時間:',is_date)

# 抓每張內文照片 如果要a裡的網址 必須先把a找出來 再用for找出href 無法一開始就找
href_list=[]
imgs=page_divs.find_all('a')
#print(imgs)
for img in imgs:
    href=img['href']
    # 如果是內文的網址 就加入list
    if href.startswith('https://i.imgur.com/'):
        href_list.append(href)
print(href_list)
print('本文照片總數量:',len(href_list))

# 下載照片
img_Dir='out8_17'
if os.path.exists(img_Dir)==False:
    os.mkdir(img_Dir)

for h in href_list:
    try:
        photo=requests.get(h)
        photo.raise_for_status()
        print('圖片下載成功',h,photo)
         # 注意大小寫是否打錯 若顯示下載成功 卻沒存進去 應該都是打錯字
         # open打開(os.path.join路徑拼接(img_Dir資料夾名稱,os.path.basename不取路徑只取檔名(h要給整個網址)),'wb'圖片要用byte寫入)
        photo_file=open(os.path.join(img_Dir,os.path.basename(h)),'wb')
        
        # 分段寫入 節省效能
        for times in photo.iter_content(10240):
            photo_file.write(times)
        photo_file.close()
    # 只要有出現失敗 表示有出錯!! 剛剛這寫成os.path.basename(photo) 所以成功失敗一起跑出!!!
    except Exception as err:
        print('圖片下載失敗')
print('--所有圖片下載完成--')
```
