```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service

url='file:///C:/Users/USER/python_web_crawler/python_orgininal_codes/ch7/h7_3.html'
driver_path=Service('D:\chromedriver.exe')
driver=webdriver.Chrome(service=driver_path)
driver.get(url)

# 以下方式都驗證相對路徑的觀念
tag1=driver.find_element(By.XPATH,'//h4') 
print(tag1.text)
tag2=driver.find_element(By.XPATH,'//body/section/h4')
print(tag2.text)
tag3=driver.find_element(By.XPATH,'//section/h4')
print(tag3.text)
tag4=driver.find_element(By.XPATH,'//body/*/h1') # h1的子節點不會印出
print(tag4.text,'\n')

# 列出第一個p元素
tag5=driver.find_element(By.XPATH,'//p')
print(tag5.text,'\n')

# 索引爬取重複的元素
tag6=driver.find_element(By.XPATH,'//section/p[1]')
print(tag6.text)
tag7=driver.find_element(By.XPATH,'//section/p[2]')
print(tag7.text,'\n')

# 列出兩個p節點的內容
tag8=driver.find_element(By.XPATH,'//section/p[@class="year"]')
print(tag8.text)
tag9=driver.find_element(By.XPATH,'//section/p[@class="price"]')
print(tag9.text,'\n')

# 列出圖片完整路徑
tag10=driver.find_element(By.XPATH,'//section/img')
print(tag10.get_attribute('src'),'\n')

# 說明textContent.innerHTML.outHTML的用法
n1=driver.find_element(By.XPATH,'//h1/em')
print('n1.text:',n1.text)
n2=driver.find_element(By.XPATH,'//h1')
print('n2.text:',n2.text)
n3=driver.find_element(By.XPATH,'//h1')
print('n3.get_attribute("textContent")只印出包含h1裡所有文字:',n3.get_attribute('textContent'))
n4=driver.find_element(By.XPATH,'//h1')
print('n4.get_attribute("innerHTML")印出包含h1裡所有文字 及子節點的標籤:',n4.get_attribute('innerHTML'))
n5=driver.find_element(By.XPATH,'//h1')
print('n5.get_attribute("outerHTML")印出包含h1裡所有物件:',n5.get_attribute('outerHTML'))

driver.close()
```

```python
# 用python填寫表單和送出 By.CLASS_NAME

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
import time

url='http://www.mcut.edu.tw/?Lang=en'
driver_path=Service('D:\chromedriver.exe')
driver=webdriver.Chrome(service=driver_path)
driver.get(url)

lists=['王永慶','郭台銘','馬雲']
for list_ in lists:
    #tag1=driver.find_element(By.CLASS_NAME,'form-control')  # 找標籤裡的class
    #tag1=driver.find_element(By.NAME,'SchKey')  # 找標籤裡的name
    tag1=driver.find_element(By.ID,'hdSchKey')  # 找標籤裡的ID
    tag1.send_keys(list_)
    time.sleep(5)
    tag1.submit()  # 將數據發送到表單後 將使用方法提交表單
    #tag1.send_keys(Keys.ENTER)  # 也可以用keys呼叫ENTER按鍵送出唷!!!!!!!!!!!!!!!!
    time.sleep(5)
    driver.find_element(By.CLASS_NAME,'form-control').clear()  # 清除標籤
    print('這裡示範用submit()提交表單:',driver.current_url)
       
driver.close()
```

```python
# 用python處理使用網頁的特殊按鍵

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import time

url='http://www.mcut.edu.tw'
driver_path=Service('D:\chromedriver.exe')
driver=webdriver.Chrome(service=driver_path)
driver.get(url)

tag1=driver.find_element(By.TAG_NAME,'body')
time.sleep(3)
tag1.send_keys(Keys.PAGE_DOWN)  # 網頁捲動到下一頁
time.sleep(3)
tag1.send_keys(Keys.END)  # 網頁捲動到最底端
time.sleep(3)
tag1.send_keys(Keys.PAGE_UP)  # 網頁捲動到上一頁
time.sleep(3)
tag1.send_keys(Keys.PAGE_UP)  # 網頁捲動到上一頁
time.sleep(3)
tag1.send_keys(Keys.HOME)  # 網頁捲動到最上端
time.sleep(3)

for i in range(7):
    tag1.send_keys(Keys.DOWN)  # 鍵盤向下鍵
    print(i)
    time.sleep(1)

driver.close()
```
