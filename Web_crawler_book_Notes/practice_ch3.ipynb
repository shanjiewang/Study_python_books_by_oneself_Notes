```python
import requests
import re

url='https://hiking.biji.co/'
try:
    response=requests.get(url)  # requests.get最好放在try後面
    response.raise_for_status()
    print('下載成功並查看原網址: %s'% response.url)  # 查看原網址
except Exception as err:
    print('網頁下載失敗 %s' % err)
    print('程式結束')

if response.status_code==requests.codes.ok:
    print('\n網址:',response.url)
#     print('\nbytes類型資料:',res.content)
    print('\n表頭內容:',response.headers)
    print('\ncookie訊息:',response.cookies)
    print('\n網頁內容:',response.text)
    print('\n網頁內容大小:',len(response.text))
    finds=input('請輸入山名:')
    
    #第一種方法
    if finds in response.text:
        print('%s 找到囉' % finds)
    else:
        print('%s 找不到' % finds)
        
    #第二種方法  
    names=re.findall(finds,response.text)
    if names != None:
        print("%s 出現 %d 次" % (finds, len(names)))
    else:
        print("%s 出現 0 次" % finds) 
else:
    print('取得網頁內容失敗')
```

```python
# urllib.request用法

import urllib.request

url = 'https://hiking.biji.co/'
headers={'user-agent': 'Mozilla/5.0 (Windows NT 6.2; Win64; x64)'
         'AppleWebKit/537.36 (KHTML, like Gecko)'
         'Chrome/98.0.4758.102 Safari/537.36'}
req=urllib.request.Request(url,headers=headers) # 偽裝成瀏覽器
response = urllib.request.urlopen(req,timeout=20)  # timeout=20預設是20 超出此時間就會跳出異常 
rr=response.read()  # read()才讀得到內容
# 使用decode解碼中文二進制
decode_obj=rr.decode('utf-8')
print('\n解碼:')
print(decode_obj)
encode_obj=decode_obj.encode()
print('\n編碼:')
print(encode_obj)
print('\n認識http.client.HTTPResponse物件屬性:')
print('版本:%s' % response.version)
print('網址:%s' % response.geturl())
print('下載:%s' % response.status)
print('表頭:')
for header in response.getheaders():
    print(header)
```

```python
# urllib.parse模組 
# 網址編解碼 quote將中文編碼成URL編碼 unquote將URL編碼解碼成中文

from urllib import parse
import webbrowser
import time

s='王山介愛爬山'
qu=parse.quote(s)
print('quote將中文編碼成URL編碼:')
print(qu)
f='%E5%8F%B0%E7%81%A3%E7%A9%8D%E9%AB%94%E9%9B%BB%E8%B7%AF%E8%A3%BD%E9%80%A0' # 台灣積體電路製造
unqu=parse.unquote(qu)
print('\nunquote將URL編碼解碼成中文:')
print(unqu)

s = '台灣積體電路製造'
# 網址是中文 也會被處理成URL編碼再傳送 伺服器端會解碼
t = 'https://zh.wikipedia.org/wiki/'+s
print('\n',t)
webbrowser.open(t)
time.sleep(5)
print('ok')
```

```python
# 解析URL

from urllib import parse

url='https://hiking.biji.co/index.php?q=trail&act=detail&id=54'
urp = parse.urlparse(url)
print(type(urp))
print(urp)
print('scheme   = ', urp.scheme)  # URL方案
print('netloc   = ', urp.netloc)  # 網路位置
print('path     = ', urp.path)    # 分層路徑
print('params   = ', urp.params)  # 最後路徑元素參數
print('query    = ', urp.query)   # 片段標示符號
print('fragment = ', urp.fragment)
```

```python
# 合成URL

from urllib import parse
import requests

scheme   =  'https'
netloc   =  'hiking.biji.co'
path     =  '/index.php'
params   =  ''
query    =  'q=trail&act=detail&id=54'
fragment =  ''

url_merge=parse.urlunparse((scheme,netloc,path,params,query,fragment))  # 要雙括號
print(url_merge)

response=requests.get(url_merge)
print('\n查看哪一種編碼:',response.encoding)

# 將json格式轉成python物件
jurl = 'https://www.httpbin.org/response-headers?freeform='
j_url=requests.get(jurl)
print('\n回傳資料格式:',j_url.headers.get('contect-type'))
print('將json格式轉成python物件:',j_url.json())

# 將網址字串轉為字典
dicts=parse.parse_qs(query)
print('\n將網址字串轉為字典',dicts)
```

```python
# urllib error異常處理

from urllib import request, error

err_url = 'https://hiking.biji.co1/'
right_url = 'https://hiking.biji.co/'
bb='http://aaa.24ht.com.tw/'

headers={'user-agent': 'Mozilla/5.0 (Windows NT 6.2; Win64; x64)'
         'AppleWebKit/537.36 (KHTML, like Gecko)'
         'Chrome/98.0.4758.102 Safari/537.36'}

# 錯誤網址 URL錯所以跳過HTTP
try:
    h=request.urlopen(err_url)
except error.HTTPError as e:
    print('錯誤代碼:',e.code)
    print('錯誤原因:',e.reason)
    print('回應的表頭:',e.headers)
except error.URLError as e:
    print('錯誤原因:',e.reason)
else:
    print('成功')
    
    
# 正確網址但錯403 HTTP錯所以跳過URL
print('\n')
try:
    h=request.urlopen(bb)
except error.HTTPError as e:
    print('錯誤代碼:',e.code)
    print('錯誤原因:',e.reason)
    print('回應的表頭:',e.headers)
except error.URLError as e:
    print('錯誤原因:',e.reason)
else:
    print('成功')
    
# 正確網址
try:
    req=request.Request(right_url,headers=headers)
    h=request.urlopen(req)
except error.HTTPError as e:
    print('錯誤代碼:',e.code)
    print('錯誤原因:',e.reason)
    print('回應的表頭:',e.headers)
except error.URLError as e:
    print('錯誤原因:',e.reason)
else:
    print('成功')
```

```python
# GET() 與 POST() 方法使用

import requests,json

# GET() params
url = 'https://www.httpbin.org/get'
ff={'gender':'M','page':'1'}
r=requests.get(url,params=ff)
print(r.url)
print(r.text)
print(r.status_code)   # 正常回應時 顯示200
print(r.reason)        # 正常回應時 顯示OK

print('~'*20)

# POST() data
headers = { 'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64)\
            AppleWebKit/537.36 (KHTML, like Gecko) Chrome/45.0.2454.101\
            Safari/537.36', }
url = 'https://www.httpbin.org/post'
ff={'gender':'M','page':'1'}
# r=requests.post(url,data=ff)
# 將ff放入data 變成"data": "{\"gender\": \"M\", \"page\": \"1\"}"
# r=requests.post(url,data=json.dumps(ff)) # dumps是 py轉json 
r=requests.post(url,json=ff,headers=headers)  #結果同上 最簡單寫法
print(r.url)
print(r.text)
print(r.status_code)   # 正常回應時 顯示200
print(r.reason)        # 正常回應時 顯示OK
```
