```python
# 使用bs4找table裡需要的連結

import time,requests,json,bs4
import pandas as pd

url='https://taipeicity.github.io/traffic_realtime/'
html=requests.get(url)
obj_soup=bs4.BeautifulSoup(html.text,'lxml')

json_url=obj_soup.find_all('table')[5].find_all('tr')[2].a['href']
print(json_url)

fn='Youbike_me.csv'
json_url=requests.get(json_url)
#print(json_url.text)
print(type(json_url.text))
youbike=json.loads(json_url.text)  #轉乘json的dict檔
# print(youbike)
print(type(youbike))

# 直接將json檔轉成df 把ybike["retVal"]內資料轉成df
df=pd.DataFrame.from_dict(youbike['retVal'],orient='index')  # orient='index'使用字典鍵當索引

df_93=df.loc[df['sna']=='吳興公車總站']  # 找出符合指定df['sna']的那列整行
print(df_93)
print()
sna=df_93['sna']
print('站點名稱:',sna)
print('更新時間:',df_93['mday'])
print('停車格數:',df_93['tot'])
print('車輛數量:',df_93['sbi'])
print('空位格數:',df_93['bemp'])

df.to_csv(fn,index=False,encoding='utf-8')  # index=False不要預設索引 json必存成utf-8檔
df
```

```python
# 使用df找出指定站點資料
import pandas as pd

fn='Youbike_me.csv'
df=pd.read_csv(fn)
df_123=df.loc[df['sna']=='榮星花園']  # 找出符合指定df['sna']的那列整行
print(df_123)
print()
print('站點名稱:',df_123['sna'])
print('更新時間:',df_123['mday'])
print('停車格數:',df_123['tot'])
print('車輛數量:',df_123['sbi'])
print('空位格數:',df_123['bemp'])

df=df[['sna','mday','tot','sbi']] # 列出需要找的行整列值 要找兩個columns以上要雙引號[['1','2']]
print(df.head())
print('-'*60)
df['車輛效率']=1-df['sbi']/df['tot']  # 可以整個表格一起計算
print(df.head())
print('-'*60)
ubike_use=df[['sna','車輛效率']]
print(ubike_use.head())
```

```python
# 直接用json找出指定站點資料
import pandas as pd
import requests,json
url='https://tcgbusfs.blob.core.windows.net/blobyoubike/YouBikeTP.json'
html=requests.get(url)
json_data=json.loads(html.text)

#ubike_df=pd.DataFrame(columns=('sna','mday','tot','sbi','bemp'))
row_id=115  # 打開csv找到索引

df=pd.DataFrame.from_dict(json_data['retVal'],orient='index')
sna=df.iloc[row_id-1,1]
mday=df.iloc[row_id-1,5]
tot=df.iloc[row_id-1,2]
sbi=df.iloc[row_id-1,3]
bemp=df.iloc[row_id-1,12]
print('站點名稱:',sna)
print('更新時間:',mday)
print('停車格數:',tot)
print('車輛數量:',sbi)
print('空位格數:',bemp)
```

```python
# 自我練習 字典轉df各種寫法
# 參考https://www.delftstack.com/zh-tw/howto/python-pandas/how-to-convert-python-dictionary-to-pandas-dataframe/
import pandas as pd

data2 = {
    '1':{
        'apple':11, 
        'banana':18}, 
    '2':{
        'apple':16, 
        'banana':12}
}
# 適合json檔轉df
df=pd.concat({k: pd.Series(v) for k,v in data2.items()},axis=0)  # axis可設定方向
df=df.reset_index()  # 先重新排列索引
df.columns=['index','A','B']  # 再給name名
print(df)

data = {
    '1':{
        'apple', 
        'banana'}, 
    '2':{
        'apple111', 
        'banana2222'}}
# 多值字典轉換成df
df=pd.DataFrame(data.values(),index=data.keys(),columns=['AAA','BBB'])
print(df)

# 鍵當name轉換成df
df=pd.DataFrame([data.values()])
df.columns=data.keys()
print(df)
```
