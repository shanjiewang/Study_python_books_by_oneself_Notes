```python
import bs4,requests,os

url='https://movies.yahoo.com.tw/movie_thisweek.html'
html=requests.get(url)
obj=bs4.BeautifulSoup(html.text,'lxml')
#print(obj)

# 本周新片總筆數
new_total_nums=obj.find('div','release_time').find('p').text
print(new_total_nums,'\n')

# 本周新片名 用find_all因為有很多部 要取得每部資料
new_names=obj.find_all('div','release_info')

new_num=0
for new in new_names:
    new_num+=1
    # find可以找完class直接給標籤名 像是find('div','class').a
    zh_name=new.find('div','release_movie_name').a.text.strip()  
    en_name=new.find('div','en').a.text.strip()
    show_date=new.find('div','release_movie_time').text
    show_date=show_date.split('：')[1].strip()  #直接取拆除字串後的索引1值
    degree=new.find('div','leveltext').span.text.strip()
    full_text=new.find('div','release_text').span.text.strip()  # 雖然頁面沒顯示全文 但因為有::before僞元素選取器 就可以抓到全文了
    print('新片編號:',new_num)
    print('中文片名:',zh_name)
    print('英文片名:',en_name)
    print('上映日期:',show_date)
    print('期待程度:',degree)
    print('影片大綱:',full_text)
    print()

# 先取得圖片連結建成list 再從list去get(圖片網址)下載圖片
# 記得如果取一個以上圖片超連結要用find_all 並且a不能直接在後接續 要用for接值
photos=obj.find_all('div','release_foto')
photo_list=[]
for photo in photos:
    # 如果標籤沒屬性 直接接續標籤名稱就好
    photo_src=photo.a.img['src']
    photo_list.append(photo_src)
#print('取得圖片連結list:',photo_list)
print('圖片數量總數:',len(photo_list))

# 建立圖片資料夾
photo_Dir='out9_5'
if os.path.exists(photo_Dir)==False:
    os.mkdir(photo_Dir)

# 下載圖片
for photo in photo_list:
    try:
        loads=requests.get(photo)
        loads.raise_for_status()
        print('圖片下載成功',photo)
        photo_file=open(os.path.join(photo_Dir,os.path.basename(photo)),'wb')
        # 會用for分開就是要慢慢寫入 所以是write(l)
        for l in loads.iter_content(10240):
            photo_file.write(l)    
        photo_file.close()  # 要記得關閉 很重要  
    except Exception as err:
        print('圖片下載失敗')
print('---所有圖片下載完成---')
```

```python
# 透過爬取兄弟結點取得海報網址

import bs4,requests,os

url='https://movies.yahoo.com.tw/movie_thisweek.html'
html=requests.get(url)
obj=bs4.BeautifulSoup(html.text,'lxml')

photo_list=[]
photo_Dir='out9_5'
if os.path.exists(photo_Dir)==False:
    os.mkdir(photo_Dir)

# 取得需要的內容
new_num=0
new_names=obj.find_all('div','release_info')
for new in new_names:
    new_num+=1
    zh_name=new.find('div','release_movie_name').a.text.strip()
    en_name=new.find('div','en').a.text.strip()
    # find_previous(兄弟結點的標籤) 透過兄弟結點取得圖片網址
    photo_links=new.find_previous_sibling('div','release_foto').a.img['src']
    # 預告片網址 如果一個標籤內很多a標籤 可使用索引取值
    notice_film=new.find('div','release_btn').find_all('a')[1]  # find_all('a')[1]['href'] 這樣寫也可以 只是有attrs檢查比較好
    # 檢查href是否存在
    if 'href' in notice_film.attrs:
        notice_film=notice_film['href']
    print('新片編號:',new_num)
    print('中文片名:',zh_name)
    print('英文片名:',en_name)
    print('海報網址:',photo_links)
    print('預告片網址:',notice_film)
    print()
    photo_list.append(photo_links)
#print(photo_list)

# 下載圖片
for photo in photo_list:
    try:
        loads=requests.get(photo)
        loads.raise_for_status()
        print('圖片下載成功',photo)
        photo_file=open(os.path.join(photo_Dir,os.path.basename(photo)),'wb')
        for l in loads.iter_content(10240):
            photo_file.write(l)
        photo_file.close()
    except Exception as err:
        print('圖片下載失敗')
print('---所有圖片下載完成---')
```

```python
# 台北.全美.預告片 排行榜一起列出 自己寫出的
import bs4,requests
import pandas as pd

url='https://movies.yahoo.com.tw/chart.html'
html=requests.get(url)
obj=bs4.BeautifulSoup(html.text,'lxml')

lists={'台北':'list1','全美':'list2','預告片':'list3'}
num_list=list(lists.keys())
pd_list=[]

num=0
for l in lists.values():
    movie_ranks=obj.find('div',id=l).find('ul','ranking_list_r').find_all('li')
    for m in movie_ranks:
        name=m.span.text.strip()
        rank=m.find('div','num').text
        print(num_list[num],'名次:',rank)
        print(num_list[num],'片名:',name)
        print() 
        pd_list.append([rank,name])
    print('-'*30,'\n')
    num+=1
print(pd_list)
file=pd.DataFrame(pd_list,columns=['名次','片名'])
file
```
