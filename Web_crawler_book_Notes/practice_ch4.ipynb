```python
# Series與 numpy轉成 DataFrame
# Series(內容,索引,name='標題')

import pandas as pd
import random

a=[1,2,3]
b=[4,5,6]
c=[7,8,9]
indexs=['as','bn','cv']
titles=['A','B','C']  

# 將list合併成numpy可以直接轉DataFrame 
print('numpy可以直接轉DataFrame:')
aa=[]
# 將list合併numpy
aa.append(a)
aa.append(b)
aa.append(c)
alld=pd.DataFrame(aa,index=indexs)  # 給索引index
alld.columns=titles  # 給標題 要放DataFrame後面
print(alld)

# 無法合併list 所以要轉series 就可concat了
# index一定要一樣 不然會出現NaN
print('\n無法合併list 轉series就可concat成DataFrame了 原本的name就是標題:')
a=pd.Series(a,indexs,name='HH')
b=pd.Series(b,indexs,name='TT')  # index一定要一樣
c=pd.Series(c,indexs,name='PP')

# 轉series就可concat成DataFrame了
alls=pd.concat([a,b,c],axis=1)  # axis=0會變一排 因為是列

# 原本的name就是標題 
print(alls)

# 給新標題 
print('\n.columns給新標題:')
alls.columns=titles  
print(alls)
```

```python
# ch4_26.py  at、iat用法 資料有新增一項 跟課本不一樣
import pandas as pd
cities = {'Country': ['China', 'China', 'Thailand', 'Japan', 'Singapore','America'],
          'Town': ['Beijing', 'Shanghai', 'Bangkok', 'Tokyo', 'Singapore','Shanghai'],
          'Population': [2000, 2300, 900, 1600, 600,500]}
df = pd.DataFrame(cities, columns=["Town", "Population"],
                  index=cities["Country"])
print(df)


# at:取得index(row)為'Japan'、columns為'Town'的項目
print('\nat:取得index(row)為"Japan"、columns為"Town"的項目:')
a = df.at['Japan', 'Town']  
print(a)

# at:取得名稱index(row)為'China'、columns為'Town'的項目,因為'China'有兩個所以回傳兩項
print('\nat:取得名稱index(row)和columns的項目,因為"China"有兩個所以回傳兩項:')
a = df.at['China', 'Town']
print(a)

# iat:取得數字位置[index(row)=2,columns=0]單一項目
print('\niat:取得數字位置[index(row)=2,columns=0]單一項目:')
a = df.iat[2, 0]
print(a)

# 超實用自我練習 這方法適用columns有很多行的 兩個以上會print出DataFrame
print('\n超實用自我練習 這方法適用columns有很多行的 兩個以上會print出DataFrame:')
a=df.iloc[:,1:2].loc['China']
a
```

```python
# ch4_26.py  loc、iloc用法

# loc:取得名稱index(row)為'Singapore'的項目
print('\nloc:取得名稱index(row)為"Singapore"的項目:')
a = df.loc['Singapore']
print(a)

# loc:取得兩個名稱以上index(row)為'Japan', 'Thailand'的項目,要雙[]
print('\nloc:取得兩個名稱以上index(row)為"Japan"和"Thailand"的項目,要雙[]:')
a = df.loc[['Japan', 'Thailand']]
print(a)

# loc:取得名稱index(row)為'China':'Thailand'、columns為'Town':'Population'的項目,最多給兩個
print('\nloc:取得名稱index(row)為"China"和"Thailand"、columns為"Town"和"Population"的項目,最多給兩個:')
a = df.loc['China':'Thailand', 'Town':'Population']
print(a)

# loc:取得columns為"Town"的項目為"Shanghai"的所有項目
print('\nloc:取得columns為"Town"的項目為"Shanghai"的所有項目:')
a = df.loc[df['Town'] == 'Shanghai']
print(a)

# iloc:取得數字位置index(row)都為1的項目
print('\niloc:取得數字位置index(row)都為1的項目')
a = df.iloc[1]
print(a)

# iloc:取得數字位置index(row)=1到3列的所有項目
print('\niloc:取得數字位置index(row)=1到3列的所有項目')
a = df.iloc[1:4]
print(a)

# iloc:取得數字位置columns=1行以前的所有項目
print('\niloc:取得數字位置columns=1行以前的所有項目')
a = df.iloc[:, :2]
print(a)

# 列出columns為'Population'的所有項目
print('\n列出columns為"Population"的所有項目')
df['Population']
```

```python
import pandas as pd
import csv

# 作品:自行用亂數-創作出表格寫入csv再拿出,然後把nan填上平均數,並求各平均值並排序
# Vicky的最高分是甚麼科目幾分,English最高分是誰,改用學號當index,字典加入現有df,相同index不同df合併,subplots畫圖
df1=pd.read_csv('sorces.csv',encoding='utf-8')
print(df1)

# 判斷是否為NaN
print('\n判斷是否為NaN:')
print(df1.isna())

# 在NaN填為0 fillna可填上指定字串
print('\n在NaN填為0 fillna可填上指定數字或字串:')
print(df1.fillna(0))

print('\ndropna會把有空值得那列整行刪除 謹慎用:')
df1=df1.dropna()
print(df1)

print('\ndropna(axis="columns")會把有空值得那行整行列刪除 謹慎用 這裡故意用df1.copy():')
df1_copy=df1.copy()
df1_copy=df1_copy.dropna(axis='columns')
print(df1_copy)

print('\nset_index()將指定列設為index:')
df1.set_index('Name',inplace=True)  # 將指定列設為index inplace：是否修改原數據
print(df1)

# 要寫完將指定列設為index後 再寫下面這幾行才不會出錯

# mean()只能求某行整列的平均 axis=1可以印出但無法放入DataFrame取代NaN
print('\nmean()只能求某行整列的平均')
df1_mean=df1.mean()
df1_mean=df1_mean.astype(int)  # 取整數
print(df1_mean)
# df1_mean=df1.mean(axis=1)  # 可以印出但無法放入DataFrame取代NaN

df1_min=df1.min()
# 將某行整列的最小值取代NaN 並轉成整數 誰叫你們要缺考哈哈 
print('\nfillna(value=df1_min)將某行整列的最小取值代NaN 並轉成整數:')
df1=df1.fillna(value=df1_min)
df1=df1.astype(int)  # astype(int)將DataFrame所有數值轉成整數
print(df1)
```

```python
# 網路範例 https://www.pythonprogramming.in/how-to-pie-chart-with-different-color-themes-in-matplotlib.html
import matplotlib.pyplot as plt
 
sizes = [12, 23, 11, 17, 19, 24, 29, 11, 12, 9, 7, 5, 3, 2, 1]
labels = ["Market %s" % i for i in sizes]
 
fig1, ax1 = plt.subplots(figsize=(5, 5))
fig1.subplots_adjust(0.3, 0, 1, 1)
 
theme = plt.get_cmap('copper')
ax1.set_prop_cycle("color", [theme(1. * i / len(sizes))
                             for i in range(len(sizes))])
 
_, _ = ax1.pie(sizes, startangle=90, radius=1800)
 
ax1.axis('equal')
 
total = sum(sizes)
plt.legend(
    loc='upper left',
    labels=['%s, %1.1f%%' % (
        l, (float(s) / total) * 100)
            for l, s in zip(labels, sizes)],
    prop={'size': 11},
    bbox_to_anchor=(0.0, 1),
    bbox_transform=fig1.transFigure
)
 
plt.show()
```

```python
# apply()介紹 https://zhuanlan.zhihu.com/p/340770847

# DataFrame 使用apply()
import numpy as np

df =pd.DataFrame([[4,9]]*3,columns = ['A','B'])
print(df)
print('np.sqrt開根號:')
print(df.apply(np.sqrt)) 

print('\n計算每列值的平均:')
print(df.apply(np.mean)) 

print('\n計算每行值的平均:')
print(df.apply(np.mean,axis=1)) 

print('\n竟然有這種寫法 A列+B列:')
df['C']=df.A+df.B  # 新寫法
# df['C']=df['A']+df['B']  
print(df['C'])

print('\n使用函數添加A列+B列 相當於上列寫法:')
def add_C(x):
    return x.A+x.B
df['C']=df.apply(add_C,axis=1)  # apply()內無法運算 所以要用函式帶入
print(df['C'])


# Series 使用apply()

print('\n列A中所有元素加1 不用apply():')
df.A =df.A +1
print(df)

print('\n列A中所有元素加1 使用apply()並傳入lambda:')
df.A = df.A.apply(lambda x:x+1)
print(df)

print('\n判斷列A 中元素是否能夠被2整除 用Yes 或No在旁邊標註:')
# lambda parameter_list: expression：lambda關鍵字,parameter_list(參數清單),expression(運算式)
df.A = df.A.apply(lambda x:str(x)+"\tYes" if x%2==0 else str(x)+"\tNo")
print(df)
```
