```python
# 小數據 可使用自己pandas方式 將台灣銀行利率表存成csv檔
# 但這種一團一起抓下來的內容 都要整理成雙list在分成每筆小list一行行寫入csv較好
import requests,bs4,csv
import pandas as pd

columns_list,table_list,tables_list,tables_list_1=[],[],[],[]

url='http://www.taiwanrate.com/'
headers={'user-agen': 'Mozilla/5.0 (Windows NT 6.2; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.75 Safari/537.36'}
html=requests.get(url,headers=headers)
html.encoding='utf-8'
obj_soup=bs4.BeautifulSoup(html.text,'lxml')

obj_left=obj_soup.find('table',id='table1').find('tr','bodytablehead').td  # 由於網頁結構關係 此為印出'銀行'2字
obj_columns=obj_soup.find('table',id='table1').find('tr','bodytablehead').find_all('a','bodytablehead')
obj_table=obj_soup.find('table',id='table1').find_all('tr',align='right')
print(obj_left.text.strip(),end=' ')
columns_list.append(obj_left.text.strip())
for obj_c in obj_columns:
    #print(obj_c.text.strip(),end=' ')  # 加end=' '就是印出不換行
    columns_list.append(obj_c.text.strip())

print(columns_list)
lens=len(columns_list)
tables_list_1.append(columns_list)  # 使用to_csv就不需要這行 
print()

# 這種一大群資料一起抓下來的 要先將所有內容加入到同一個list
for table in obj_table: 
    for td in table:
        tds=td.text.strip()
        #print(tds)
        table_list.append(tds)
#print(table_list)
print(len(table_list))

# 再依照list每筆內容的長度 將其分開成每筆小list 再加入一個新大list變雙list 也是大數據要一行行寫入csv的重要方式 但這裡做法是一次存成csv
for r in range(0,len(table_list),lens):
    table_1=table_list[r:r+lens]   # 分割list
    #print(table_1)
    tables_list_1.append(table_1)
    tables_list.append(table_1)   # 使用to_csv的list
print(tables_list_1)    

df_file='out_13_me.csv'
counts=0

# 大數據要用writer及for writerow一筆筆寫入csv 並用try except確認-----
with open(df_file,'w',newline='',encoding='utf-8') as csv_file:   # 記得存csv時都給encoding='utf-8'比較保險
    csv_writer=csv.writer(csv_file)
    for row in tables_list_1:
        try:
            csv_writer.writerow(row)
            counts+=1
            print('寫入成功',counts)
        except Exception:
            print('寫入失敗')
            pass
print('全部寫入完成')   

# # 小數據再用to_csv方式寫入csv-----
# df=pd.DataFrame(tables_list,columns=columns_list)
# df.to_csv(df_file,index=False)
# df
```

```python
# 將基金資料存成csv
import requests,bs4,csv
import pandas as pd

headers={'user-agent': 'Mozilla/5.0 (Windows NT 6.2; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.75 Safari/537.36'}
url='https://www.moneydj.com/funddj/ya/yp401000.djhtm'
html=requests.get(url,headers=headers)
obj_soup=bs4.BeautifulSoup(html.text,'lxml')

columns_list,contect_list,contect_list_1=[],[],[]

obj_tr=obj_soup.find('table',id='oMainTable').find('tr',id='oScrollMenu').find_all('th')
obj_tbody=obj_soup.find('table',id='oMainTable').find('tbody').find_all('tr')
# 標題
for tr in obj_tr:
    print(tr.text.strip())
    columns_list.append(tr.text.strip())
contect_list_1.append(columns_list)

# 內容
for tb in obj_tbody:
    txts=tb.text.strip() # 要觀察網頁寫法 這裡是用\n分隔
    contect=txts.split('\n')  # 變成list
    #print(contect)
    contect_list.append(contect)   # to_csv
    contect_list_1.append(contect)   # writerow
    
df_file='out_13_1_me.csv'

# 用大數據一筆筆匯入方式
counts=0
with open(df_file,'w',newline='') as csv_file:
    csv_writer=csv.writer(csv_file)
    for r in contect_list_1:
        counts+=1
        try:
            csv_writer.writerow(r)
            print('寫入成功',counts)
        except Exception:
            print('寫入失敗')
print('全部寫入完成')
    
# # 用pandas方式轉成雙list
# df=pd.DataFrame(contect_list,columns=columns_list)
# df.set_index('排名',inplace=True)
# df.to_csv(df_file,encoding='utf-8')
# df
```
