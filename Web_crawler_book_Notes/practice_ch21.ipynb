```python
# 使用By.xpath定位元素 位置開頭要//

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import time

url_yahoo='https://tw.bid.yahoo.com/search/auction/product?'
url_product='kw=薩爾達傳說&p=薩爾達傳說&sort=-ptime'
url=url_yahoo+url_product
driver_path=Service('D:/chromedriver.exe')
driver=webdriver.Chrome(service=driver_path)

# time.sleep(10) 恰好暫停代碼執行 10 秒。
# driver.implicitly_wait(10) 最多等待 10 秒以等待元素的存在。如果在 2 秒後找到，則將繼續執行代碼，而無需等待更多 8 秒。

driver.implicitly_wait(5) # 通\常放get.url前面
driver.get(url)

# 商品連結
link_path="//ul[@class='gridList']/li/a" 
product_links=driver.find_elements(By.XPATH,link_path)  # find_elements如果有s表現找很多 後面不可有標籤

# 商品名稱
title_path="//span[@class='BaseGridItem__title___2HWui']"
product_titles=driver.find_elements(By.XPATH,title_path)

# 商品價格
price_path="//span[@class='BaseGridItem__price___31jkj']"
product_prices=driver.find_elements(By.XPATH,price_path)

print('商品數量:',len(product_links))
print('-'*50)

for title,link,price in zip(product_titles,product_links,product_prices):
    print('商品名稱:',title.get_attribute('innerHTML'))
    print('商品連結:',link.get_attribute('href'))
    print('商品價錢:',price.text.strip())
    print('-'*50)
    
# 可參考get_attribute的幾種用法 https://blog.csdn.net/gaimechen/article/details/81947879

driver.close()
```

```python
# 練習用selenium.By.CSS_SELECTOR取得圖片連結

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import time

url_yahoo='https://tw.bid.yahoo.com/search/auction/product?'
url_product='kw=薩爾達傳說&p=薩爾達傳說&sort=-ptime'
url=url_yahoo+url_product
driver_path=Service('D:/chromedriver.exe')
driver=webdriver.Chrome(service=driver_path)

# time.sleep(10) 恰好暫停代碼執行 10 秒。
# driver.implicitly_wait(10) 最多等待 10 秒以等待元素的存在。如果在 2 秒後找到，則將繼續執行代碼，而無需等待更多 8 秒。

driver.implicitly_wait(5) # 通\常放get.url前面
driver.get(url)

products=driver.find_elements(By.CSS_SELECTOR,'.BaseGridItem__grid___2wuJ7 a')
print('用索引找第一個網址:',products[0].get_attribute('href'))
print()
for product in products:
    print(product.get_attribute('href'))

driver.close()
```

```python
# 練習用bs4取得圖片連結
import bs4,requests

url_yahoo='https://tw.bid.yahoo.com/search/auction/product?'
url_product='kw=薩爾達傳說&p=薩爾達傳說&sort=-ptime'
url=url_yahoo+url_product
response=requests.get(url)
obj_soup=bs4.BeautifulSoup(response.text,'lxml')

products_link=obj_soup.find_all('li','BaseGridItem__grid___2wuJ7')
for product in products_link:
    print(product.a['href'])
```
