```python
# Python 網路連線程式、公開資料串接
# 參考考https://www.youtube.com/watch?v=sUzR3QVBKIo
import urllib.request as request

url='https://www.ntu.edu.tw/'
with request.urlopen(url) as response:
    data=response.read().decode('utf-8')  # 讀取網站原始碼並解碼
print(data)
```

```python
import urllib.request as request
import json

url='https://data.taipei/api/v1/dataset/296acfa2-5d93-4706-ad58-e83cc951863c?scope=resourceAquire&limit=50'
with request.urlopen(url) as response:
    data=json.load(response)  # 讀取網站原始碼並解碼
lists=data["result"]["results"]
print(type(lists))
with open('data.txt','w',newline='',encoding='utf-8') as file:
    for list_ in lists:
        print(list_["公司名稱"])
        file.write(list_["公司名稱"]+'\n') # 加\n才會分行 不會擠成一團
        
```

```python
import json,requests
import pandas as pd

url='https://data.taipei/api/v1/dataset/296acfa2-5d93-4706-ad58-e83cc951863c?scope=resourceAquire&limit=50'
response=requests.get(url)
data=json.loads(response.text)
print(type(data))

lists=data["result"]["results"]
with open('data_me.txt','w',newline='',encoding='utf-8') as file:
    for lists_ in lists:
        print(lists_["公司名稱"])
        file.write(lists_["公司名稱"]+'\n')
```
