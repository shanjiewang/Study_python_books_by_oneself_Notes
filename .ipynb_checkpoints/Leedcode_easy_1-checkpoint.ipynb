```python
import random

lists=[]
for i in range(1,100000):
    num=random.randrange(-100000, 100000)
    lists.append(num)
# print(lists)

def Max_int_fn(lists):
    A = set(lists)
    max_num = 1
    while max_num in A:
        max_num += 1
    return max_num

Max_int_fn(lists)
```

```python
A=[6,6,5,6,5,6,5]

# A=[]
# for i in range(0,100000):
#     num=random.randrange(-100000, 100000)
#     A.append(num)
def solution(A):
    lists=[]
    count=0
    for i,o in enumerate(A):
        if i!=len(A)-1:
            a=A[i]
            b=A[i+1]
            s=a+b     
            if s%2==0:
                a='Y'
            else:
                a='N'
        else:
            s=A[i]+A[0]       
            if s%2==0:
                a='Y'
            else:
                a='N'
        lists.append(a)  
    for i in range(0,len(lists)):
        if i!=len(A)-1:
            if i%2==0:
                if lists[i]=='Y' or lists[i+1]=='Y':
                    count+=1
        else:
            if lists[i-1]=='N':
                if lists[i]=='Y' or lists[0]=='Y':
                    count+=1
    return count

solution(A)
```

```python
X='.....xx...x.x.x....x.xxxxx.'

def solution(X):
    count=0
    for i in range(len(X)-1):
        if X[i]=='x': 
            if X[i+1]=='.':
                print('nn')
                count+=1
            else:
                print('mm')

    return count
solution(X)
```
