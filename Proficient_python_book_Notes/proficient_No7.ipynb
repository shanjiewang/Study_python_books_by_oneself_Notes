```python
## 再看一個例子：「有一個 a 串列，接著要建立一個 b 串列，b 串列每個內容項目是 a 串列的最大值減去其他項目的值」，這時使用串列生成式，整個程式碼就會變得非常簡潔。

print('\n串列生成式:')
# 單純寫法
a = [10,20,30,40,50,60,70,80,90]
b = []
for i in a:
    b.append(max(a) - i)     # 用 a 的最大值減去每個項目
print(b)                  

# 使用串列生成式
a = [10,20,30,40,50,60,70,80,90]
b = [max(a)-i for i in a]
print(b)                  


## 如果需要兩層 for 迴圈才能生成的串列，同樣也能使用串列生成式來產生。

# 單純寫法
# 將兩層 for 迴圈的 i 和 j 加在一起，變成新串列的項目
print('\n將兩層 for 迴圈的 i 和 j 加在一起，變成新串列的項目:')
a = []
for i in 'abcd':
    for j in range(1,4):
        a.append(i + str(j))
print(a)        

# 使用串列生成式
# 兩個 for 迴圈分別產生 i 和 j
a = [i + str(j) for i in 'abcd' for j in range(1, 4)]
print(a)    


## 串列生成式也可以加入 Python 的內建函式，針對產生的項目做處理，下面的程式，只要透過一行串列產生式，就能取出二維陣列裡的最小值。

print('\n加入 Python 的內建函式:')
# 單純寫法
a = [[100, 200, 300, 400, -2], [100, 200, 500, 2, 1]]
b = []
for i in a:
    b.append(min(i))  # 將二維串列中每個串列裡的最小值取出，變成新的串列
print(min(b))       # 印出新的串列裡的最小值

# 使用串列生成式
a = [[100, 200, 300, 400, -2], [100, 200, 500, 2, 1]]
print(min([min(i) for i in a]))   


## 如果將 if 放在 for 的前方，就必須加上 else ( 三元運算式 ( 條件運算式 ) )，下方的例子，會將偶數的項目保留，奇數項目替換成 100。

print('\n如果將 if 放在 for 的前方，就必須加上 else(三元運算式(條件運算式)):')
a = []
for i in range(1,10):
    if i%2 == 0:
        a.append(i)   # 取出偶數放入變數 a
    else:
        a.append(100) # 如果是奇數，將 100 放入變數 a
print(a)          

a = [i if i%2==0 else 100 for i in range(1, 10)]
print(a)  
```
