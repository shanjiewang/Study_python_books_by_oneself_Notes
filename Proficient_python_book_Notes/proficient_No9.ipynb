```python
#### 函式 ####

# 最基本
print('最基本:')
def test1():
    return True
print(test1())
if test1():
    print('test1')
else:
    print('NO test1')

print('\n基本:')
def colors(c):
    if c=='red':
        return 'RED!!'
    elif c=='blue':
        return 'BLUE!!'
    else:
        return 'NO COLOR!!'

com=colors('red')
print(com)

# 此欄位是否有資料?? 沒有就填上None
print('\n沒資料填上None:')
test2=['1','','oo']
for i in range(len(test2)):
    if test2[i] !='':
        print(f'yes: {i}... {test2[i]}')
    else:
        test2[i]='None'
        print(f'no: {i}... {test2[i]}')
print(test2)

# 測試各物件的空值是否是None
print('\n測試各物件的空值是否是None:')
def whatis(obj):
    if obj is None:
        print(f'{obj} in None')
    elif obj:
        print(f'{obj} in True')
    else:
        print(f'{obj} in False')

whatis("")    
# whatis("") 卻是None???  
whatis(0.0)
whatis([])
whatis({})
whatis(['0'])
```

```python
# 關鍵字引數 與 指定預設參數

print('\n關鍵字引數是找關鍵字 位置亂掉也沒差  指定預設參數:')
def meun1(A,B='###',C='123',D='您未給值喔'):
    print({'A':A,'B':B,'C':C,'D':D})
    return {'A':A,'B':B,'C':C,'D':D}

# 假如有指定預設參數 呼叫函式時沒給引數也ok
meun1(B='bbbbb',A='aaaa',C='cccccc')

print('\n將資料一項一項加入list:')
qq=[]
def buggy(arg,result=[]):
    result.append(arg)
    qq.append(arg)
    print(result,'...',qq)
buggy('a')
buggy('')
buggy('b')


print('\n將資料每次加入list 不累加:')
def ok1(arg,result=None):
    if result is None:
        result=[]
    result.append(arg)
    print(result,'...')
ok1('a')
ok1('')
ok1('b')
```

```python
print('\n*蒐集位置引數:')
def bom(b1,b2,*bn):
    print(f'b1:{b1}...b2:{b2}...bn:{bn}')
bom('aaaaa','bbbbbb','cccccc','ddddddd','eeeeeee')

print('\n**蒐集關鍵字引數:')
def bom2(**bn):
    print(f'bn:{bn}')
bom2(A='aaaaa',B='bbbbb',Z='zzzzz')

print('\n*純關鍵字引數有起始:')
def bom3(data, *,start=2,end=100):
    for value in (data[start:end]):
        print(f'value:{value}')
out=['a','b','c','d','e']
bom3(out)

print('\n*可變與不可變引數 不好的做法:')
outside=['one','fine','day']
def mangle(arg):
    arg[1]='changeeeeeee'
mangle(outside)
print(outside)

print('\n*可讀性很重要:')
def echo(thing,check):
    '''
    好多字
    這裡是
    註解區
    '''
    if check:
        print(thing)
    else:
        print('Nothing')
echo('我是False',0)   
echo('印出True的我',2) 

# 印出函式docstrings文檔字符串
help(echo)
help(echo.__doc__)
```

```python
## 函式是一級公民 從函式的引數呼叫函式 ##

def answer():
    print(42)
def answerFn(func):
    func()   
answerFn(answer)
print(type(answerFn))

# 從函式的引數呼叫函式
def add_args(arg1,arg2):
    print('多層函式:',arg1+arg2)

def big_add_args(add_args,arg1,arg2):
    add_args(arg1,arg2)
    
big_add_args(add_args,5,6)

# 合併*arg和**kwargs作法
def run(*arg,**kwargs):   
    print(f'*arg: {sum(arg)}')
    print(f'*kwarg: {kwargs}')   
run(1,2,5,A=2,B=2)

# 引數是函式 run1呼叫func func回傳內容
def func(*arg): 
    print('引數是函式:',sum(*arg))
def run1(func,*arg):   
    return func(arg)  
run1(func,1,2,5,9)
```

```python
### 內部函式 ###

def out(a,b):
    def inner(c,d):
        return c+d
    return a+b
print('內部函式數字運算:',out(97,99))

# 內部函式文字顯示 用%s帶入
def out1(aa):
    def inner1(bb):
        return "I say:'%s'" % bb
    return inner1(aa)
print('內部函式文字顯示:',out1('HAHA'))

# closure
def out2(aa):
    def inin():
        return "I say:'%s'" % aa
    return inin

# 讓函式=變數 這變數就是個函式 也要用函式方式執行
a=out2('JLKJL')
b=out2('HUOLJFTGDFHJKGJHKJHLKJHL')
print(a(),",",b())
```

```python
### 匿名函式 ###

stairs=['one','two','three','two']
def story1(words,func):
    for i in words:
        print(func(i))
        
def story2(words):
    return words.upper()+'!'
story1(stairs,story2)

# 使用lambda取代def 適合用在走許多個小型函式時 算是暫時性函式
print('\n匿名函式:')
story1(stairs,lambda ss:ss.upper()+'!!')
```

```python
### 產生器函式 ###

# 把所有整數加總
print(f'把所有整數加總: {sum(range(1,101))}')

# 自己製作產生器
def my_range(first=0,last=10,step=1):
    num=first
    while num<last:
        yield num
        num=(num+step)*2
    else:
        print('Done!!')
        
ranger=my_range(1,12)
print(f'它是個一般函式: {my_range}')
print(f'它是個產生器物件: {ranger}')

print('\n自己寫產生器函式 但只能執行一次 因為他會記住上次的位置:')
for x in ranger:
    print(x)

# 跑不動了
for y in ranger:
    print('y:',y)
    
# 產生器生成式
print('\n產生器生成式:')
genobj=(pair for pair in zip(['a','b'],['1','2']))
for thing in genobj:
    print(thing)

print('\n測試zip:')
for i in zip(['q','f','u'],['1','2','3']):
    print(i)
```

```python
### 裝飾器 ###
# Python 需要這個裝飾器！
# 有時你想要修改既有的函式，但不想更改它的原始碼…………

# 假設原本的函式
def document_it(func):
    def new_function(*args,**kwargs):
        print('Running function:',func.__name__)
        print('Positional arguments:',args)
        print('keyword arguments:',kwargs)
        result=func(*args,**kwargs)
        print('Result:',result)
        return result
    return new_function

def add_ints(a,b):
    return a+b

# 手動裝飾器 給變數指派
print('手動裝飾器:')
ooo=document_it(add_ints)
ooo(3,5)

# 自動裝飾器 在新函式上面加@document_it
print('\n自動裝飾器:')
@document_it
def add_ints1(a,b):
    return a+b
add_ints1(3,8)

def sq_it(func):
    def inFn(*args,**kwargs):
        ruselt=func(*args,**kwargs)
        return ruselt*ruselt
    return inFn

print('\n自動裝飾器 2個函式以上:')
@document_it
@sq_it
def add_ints1(a,b):
    return a+b
add_ints1(3,8)
```

```python
#### 裝飾器的定義 #####
def out(func):                        # 外圍函式
    def inter(*args, **kwargs):       # 內部函式
        return func(*args, **kwargs)
    return inter                      # 外圍函式返回內部函式

# 實做簡單的裝飾器 在函式前加@
print('自製簡單裝置器:')
def check_str(func):
    def inner():
        print('Python')
        return func()
    return inner

@check_str
def say():
    print('World')    
say()

# 較複雜的裝飾器
print('\n進階一些的裝置器:')
def check_str1(func):
    def inner(*args, **kwargs):
        result = func(*args, **kwargs)
        if result == 'Python':
            return '%s, World' % result
        else:
            return '%s, Nice to meet you' % result
    return inner

@check_str1
def test(data):
    return data

result1 = test('RRRRRRRRRR')
print(result1)
result1 = test('Python')
print(result1)

# 完全自己寫出的裝置器
print('\n完全自己寫出的裝置器:')
def outfn(func):
    def inner(*args,**kwargs):
        result=func(*args,**kwargs)
        return result**2
    return inner

@outfn
def hh(a,b):
    return a+b

hh(2,4)


裝飾器也有參數?
如果想要在裝飾器上輸入參數該怎麼辦呢? 請看下列例子：
def check_str(msg):
    def outer(func):
        def inner(*args, **kwargs):
            print(msg)
            return func(*args, **kwargs)
        return inner
    return outer
@check_str(‘Hello’)
def say(msg):
    print(msg)
say(‘World’)
結果將顯示如下︰
Hello
World
```

```python
#### 名稱空間與作用域 ####

animal='fruitbat'
def print_global():
    animal='changess'  #在取得前可以更改 只是修改成同名的函式內區域變數 外面一樣不變
    print('inside:',animal)
    #animal='HHHHHH'  # 在函式內已取得全域變數後不可更改他
print('outside:',animal) # 用id看會發現裡外位置不同
print_global()

# 使用global在其他函式內也可以用
print('\nGlobal:')
def print_global1():
    global animal1
    animal1='changess'  
    print('inside1:',animal1) 
    
def print_global2(): 
    print('inside2:',animal1) # 抓得到
    
print_global1()
print_global2()

print('\nlocals與globals讀取名稱空間內容:')
animal='fruitbat123'
def change_local():
    animal='裡面的animal changess123'  #在取得前可以更改 只是修改成同名的函式內區域變數 外面一樣不變
    print('local:',locals())
print(animal)
change_local()
#print('global:',globals())  # 太長了所以先註解  __name__=__main__

# 在名稱內使用_與__
print('\n在名稱內使用_與__:')
def qq():
    print(f'function name: {qq.__name__}')
    print('docstring is:',qq.__doc__)
    #return qq()  # 函式內不可呼叫自己 打開會出錯
qq()

### 遞迴此產生器函式處理不平整資料時很方便 例如雜亂的串列 ###
print('\n遞迴此產生器函式用於處理不平整資料:')
def flatten(lol):
    for item in lol:
        if isinstance(item,list):
            for subitem in flatten(item):
                yield subitem
        else:
            yield item
lol=[1,2,[3,4,5],[6,[7,8,9],[]]]
print('雜亂資料:',lol)
print('透過產生器函式處理後:',list(flatten(lol)))

# 簡化版 
print('\n簡化版:')
def flatten(lol):
    for item in lol:
        if isinstance(item,list):           
            yield from flatten(item)                 
        else:
            yield item  # 在這行變正常

print('雜亂資料:',lol)
print('透過簡化版產生器函式處理後:',list(flatten(lol)))
```
